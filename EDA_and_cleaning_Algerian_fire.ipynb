```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv(r'd:\Practice\Pandas\archive\Algerian_forest_fires_dataset.csv')
df.head()
```

```python
df.info()
```

```python
#Dataset
df[df.isnull().any(axis =1)]
```

```python
# Dataset is divided into two parts
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
```

```python
df.info()
```

```python
df['Region'] = df['Region'].astype(int)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df = df.drop(122).reset_index(drop = True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
##fix spaces in columns names
df.columns = df.columns.str.strip()
df.columns
```

```python
#Change the required columns as integer data type
df[['day', 'month', 'year','Temperature','RH', 'Ws',]] = df[['day', 'month', 'year','Temperature','RH', 'Ws',]].astype(int)
```

```python
#####Cahnging the other columns to float
object = [features for features in df.columns if df[features].dtype=='O']
```

```python
for i in object:
    if i!='Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
#Lets save the cleaned dataset
df.to_csv('Algerian_cleaned_dataset.csv',index = False)
```

## Exploratory Data Analysis

```python
df_copy =  df.copy()
```

```python
#Dropping columns
df_copy.drop(['day','month','year'], axis =1, inplace = True)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
#Plotting density plot for all features
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-whitegrid")
df_copy.hist(bins = 50, figsize=(20,15))
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Apply Seaborn style
sns.set(style="whitegrid")

# Plotting density plot for all features
plt.figure(figsize=(20, 15))  # Adjust the figure size as needed

# Loop through each column and plot the density plot
for i, column in enumerate(df_copy.columns, 1):
    plt.subplot((len(df_copy.columns) // 3) + 1, 3, i)  # Create a grid layout
    sns.kdeplot(df_copy[column], shade=True)
    plt.title(f'Density Plot for {column}', fontsize=12)

plt.tight_layout()
plt.show()
```
